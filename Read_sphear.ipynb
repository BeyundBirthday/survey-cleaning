```python
most_common_apps = df['Current Apps'].value_counts().index[:2]
most_common_rewards = df['Preferred Rewards'].value_counts().index[:4]
```

```python
app_mode = df['Current Apps'].mode()[0]
reward_mode = df['Preferred Rewards'].mode()[0]
```

```python
df['Current Apps'] = df['Current Apps'].apply(lambda x: x if x in most_common_apps else app_mode)
df['Preferred Rewards'] = df['Preferred Rewards'].apply(lambda x: x if x in most_common_rewards else reward_mode)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

background_color = '#F4F6F7'  
text_color = '#2E4053'  

palette = {
    'Yes': '#9B59B6',  
    'No': '#3498DB',  
    'Sometimes': '#D5DBDB'  
}

object_columns = df.select_dtypes(include=['object']).columns

num_cols = len(object_columns)
cols = 3
rows = (num_cols // cols) + (num_cols % cols > 0)

fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 9))
axes = axes.flatten()

for i, col in enumerate(object_columns):
    sns.countplot(ax=axes[i], x=col, data=df, hue='Read Books', palette=palette)
    axes[i].set_title(f'{col}', fontdict={'font': 'Serif', 'color': text_color, 'weight': 'bold', 'size': 16})
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    
    axes[i].tick_params(axis='x', rotation=45, labelsize=12)
    axes[i].tick_params(axis='y', labelsize=12)
    
    axes[i].set_facecolor(background_color)

for j in range(i + 1, len(axes)):
    axes[j].axis('off')

fig.patch.set_facecolor(background_color)
plt.tight_layout()
plt.savefig('reading_app_plots.png', dpi=100, bbox_inches='tight')
plt.show()
```

```python
text_color = '#2E4053'
background_color = '#F4F6F7'

features = df.select_dtypes(include=['object']).columns.tolist()

num_features = len(features)
cols = 3
rows = (num_features // cols) + (num_features % cols > 0)

fig = plt.figure(figsize=(18, rows * 6), dpi=100)
gs = fig.add_gridspec(rows, cols)
gs.update(wspace=0.3, hspace=0.4)

axes = [fig.add_subplot(gs[i, j]) for i in range(rows) for j in range(cols)]

colormap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["#D0E1F9", "#1F77B4"], N=256)

for i, feature in enumerate(features):
    ax = axes[i]
    crosstab = pd.crosstab(df[feature], df['Read Books'], normalize='index')
    sns.heatmap(ax=ax, data=crosstab, linewidths=0.5, cmap=colormap,
                annot=True, fmt='1.0%', annot_kws={"fontsize":12}, alpha=0.9)
    
    ax.set_title(f'Distribution of {feature}', fontdict={'font':'Serif', 'color': text_color, 'weight': 'bold', 'size': 16})
    ax.set_xlabel('Read Books', fontdict={'font':'Serif', 'color': text_color, 'size': 12})
    ax.set_ylabel('Percentage', fontdict={'font':'Serif', 'color': text_color, 'size': 12})
    
    ax.set_xticklabels(df['Read Books'].unique(), fontdict={'font':'Serif', 'color': text_color, 'size': 10})
    ax.set_yticklabels(ax.get_yticklabels(), fontdict={'font':'Serif', 'color': text_color, 'size': 10})

for j in range(len(features), len(axes)):
    axes[j].axis('off')

fig.patch.set_facecolor(background_color)
for ax in axes:
    ax.set_facecolor(background_color)
plt.savefig('plot9.png', dpi=300, bbox_inches='tight')
plt.show()
```
